```python
%matplotlib nbagg
import numpy as np
import sys, copy
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib.widgets import RectangleSelector
from matplotlib.patches import Rectangle
# Import more modules you need
```

```python
outputh = 1800
outputw = 2700

import random

def split(rect, vertical):
    area = rect["w"]*rect["h"]
    if (area < min_area):
        return [rect]
    if (rect["w"] < 400 or rect["h"] < 400):
        return [rect]
    elif (vertical):
        w_new = random.randint(int(0.4*rect["w"]), int(0.6*rect["w"]))
        rect1 = {"x": rect["x"], "y": rect["y"], "w": w_new, "h": rect["h"]}
        if (random.random() < 0.7):
            new_vertical = not vertical
        else:
            new_vertical = vertical
        rectangles1 = split(rect1, new_vertical)
        rect2 = {"x": rect["x"] + w_new, "y": rect["y"], "w": rect["w"] - w_new, "h": rect["h"]}
        if (random.random() < 0.7):
            new_vertical = not vertical
        else:
            new_vertical = vertical
        rectangles2 = split(rect2, new_vertical)
    else:
        h_new = random.randint(int(0.4*rect["h"]), int(0.6*rect["h"]))
        rect1 = {"x": rect["x"], "y": rect["y"], "w": rect["w"], "h": h_new}
        if (random.random() < 0.7):
            new_vertical = not vertical
        else:
            new_vertical = vertical
        rectangles1 = split(rect1, not vertical)
        rect2 = {"x": rect["x"], "y": rect["y"] + h_new, "w": rect["w"], "h": rect["h"] - h_new}
        if (random.random() < 0.7):
            new_vertical = not vertical
        else:
            new_vertical = vertical
        rectangles2 = split(rect2, new_vertical)

    rectangles1.extend(rectangles2)
    rectangles1.append(rect)
    return rectangles1



width = 2700
height = 1800
min_area = 400000

rect = {"x": 0, "y": 0, "w": width, "h": height}
rects = split(rect, True)
# rects = [{'x':0, 'y':0, 'w':outputw/2, 'h':outputh/2}, {'x':outputw/2, 'y':0, 'w':outputw/2, 'h':outputh/2}]
fig,ax = plt.subplots()
currentAxis = plt.gca()
for rect in rects:
    currentAxis.add_patch(Rectangle((rect['x'], rect['y']), rect['w'], rect['h'],
                      alpha=1, facecolor='none'))
plt.axis([0, outputw, outputh, 0])
```

```python
import os
import re
direcontents = os.listdir('./images')
imagels = []

def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])

for image in direcontents:
    tmp = mpimg.imread('./images/' + image)
    if len(tmp.shape) > 2: # color image
        tmp = rgb2gray(tmp)
    imagels.append(tmp)

def imgratio_key(img):
    h = img.shape[0]
    w = img.shape[1]
    return w/h

sortedimages = sorted(imagels, key=imgratio_key)
        
```
